```python
from pyspark.sql.functions import monotonically_increasing_id

dfs1 = dfs.withColumn("id", monotonically_increasing_id())
dfs1.show()

df3 = df2.withColumn( "id", monotonically_increasing_id())
df3.show()

df_main = dfs1.join(df3,dfs1.id == df3.id, how='inner')
df_main.show()
```
