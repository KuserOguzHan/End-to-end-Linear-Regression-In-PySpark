##1.3 ML-2

```python
df2 = df1.selectExpr("memory","ram","screen_size","power","front_camera","rc1","rc3","rc5","rc7","price as label")
```

```python
df2.toPandas().head()
```

```python
numerical_cols = ["memory","ram","screen_size","power","front_camera","rc1","rc3","rc5","rc7"]
```

```python
label = ["label"]
```

```python
from pyspark.ml.feature import VectorAssembler
vector_assembler = VectorAssembler() \
.setInputCols(numerical_cols) \
.setOutputCol("features")
```

```python
from pyspark.ml.regression import LinearRegression
lr_obj = LinearRegression() \
.setLabelCol("label") \
.setFeaturesCol("features") 
```

```python
from pyspark.ml import Pipeline
pipeline_obj = Pipeline() \
.setStages([vector_assembler, lr_obj])
```

```python
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
```

```python
param_grid = ParamGridBuilder() \
.addGrid(lr_obj.aggregationDepth, [2,5]) \
.addGrid(lr_obj.elasticNetParam, [0.0, 0.2, 0.7]) \
.addGrid(lr_obj.epsilon, [1.35, 1.55]) \
.addGrid(lr_obj.maxIter, [10, 20]) \
.addGrid(lr_obj.regParam, [0.00, 0.01, 0.05]) \
.addGrid(lr_obj.solver, ["auto", "normal", "l-bfgs"]) \
.addGrid(lr_obj.tol, [1.0E-6, 1.0E-4]) \
.build()
```

```python
cv = CrossValidator() \
.setEstimator(pipeline_obj) \
.setEvaluator(RegressionEvaluator()) \
.setEstimatorParamMaps(param_grid) \
.setNumFolds(5) \
.setParallelism(2)
```

```python
df_train, df_test = df2.randomSplit([0.8, 0.2], seed=142)
```

```python
df_train.cache()
df_test.cache()
```

```python
cv_model = cv.fit(df_train)
```

```python
cv_model.transform(df2).select("label","prediction").toPandas().head()
```

```python
# En iyi pipeline model
best_model = cv_model.bestModel
```

```python
# En iyi pipeline model içinden lr modeli almak
lr1_model = best_model.stages[-1]
```
